```python
df_country = pd.read_excel('Country-Code.xlsx')
df_country.head()
```

```python
#combine countrycode column Country-Code.xlsx to zomato.csv dataset 
final_df = pd.merge(df,df_country,on='Country Code', how='left')
```

```python
final_df.head(2)
```

```python
## to check Data types
final_df.dtypes
```

```python
final_df.columns
```

```python
# lets find out how many diffrent countries with respect to there country  
final_df.Country.value_counts()
```

```python
country_names = final_df.Country.value_counts().index
```

```python
country_names
```

```python
country_val = final_df.Country.value_counts().values
```

```python
country_val
```

```python
## Pie Chart
plt.pie(country_val, labels = country_names)
```

```python
## Pie Chart - Top-3 countries that uses Zomato
plt.pie(country_val[:3], labels = country_names[:3])
```

```python
# using one property of pie chart  percentage 
plt.pie(country_val[:3], labels = country_names[:3], autopct = '%1.2f%%')
```

Observation: Zomato maximum records or transactions are from India
After that USA and then the United Kingdoms.

```python
## See more obsartvation 
# create dataframe using groupby
final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text'])
```

```python
final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size()
```

```python
# Convert this into data frame using reset_index() 
final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index()
```

```python
ratings = final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns = {0:'Rating Count'})
```

```python
ratings
```

## Conclutions/Observations
    1. when Ratings is between 4.5 to 4.9 ---->Excellent
    2. when Ratings are between 4.0 to 4.4 ----> Very Good
    3. when Ratings is between 3.5 to 3.9 ----> Good
    4. when Ratings is between 3.0 to 3.4 -----> Average
    5. when Ratings is between 2.5 to 2.9 -----> Average
    6. when Ratings is between 2.0 to 2.4 -----> poor

```python
ratings.head()
```

```python
# using bar plot to observetion
import matplotlib
matplotlib.rcParams['figure.figsize'] = (12,6)
sns.barplot(x="Aggregate rating", y="Rating Count", data= ratings)
```

```python
sns.barplot(x="Aggregate rating", y="Rating Count", hue = 'Rating color', data= ratings)
```

```python
## Change the color using palette
sns.barplot(x="Aggregate rating", y="Rating Count", hue = 'Rating color', data= ratings, palette = ['white', 'red', 'orange', 'yellow', 'green', 'green' ])
```
