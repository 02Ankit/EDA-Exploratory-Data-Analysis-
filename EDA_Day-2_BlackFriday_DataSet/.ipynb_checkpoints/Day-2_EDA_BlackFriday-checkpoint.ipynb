## Problem Statement
A retail company “ABC Private Limited” wants to understand the customer purchase behaviour (specifically, purchase amount) against various products of different categories. They have shared purchase summary of various customers for selected high volume products from last month. The data set also contains customer demographics (age, gender, marital status, city_type, stay_in_current_city), product details (product_id and product category) and Total purchase_amount from last month.

Now, they want to build a model to predict the purchase amount of customer against various products which will help them to create personalized offer for customers against different products.

```python
# importing the dataset
df_train = pd.read_csv('blackFriday_train.csv')
df_train.head()
```

Whenever we have test and train datasets we have to combine both dataset to perform as well as in kegel compitition always remember 

```python
## importing the test data 
df_test = pd.read_csv('blackFriday_test.csv')
df_test.head()
```

```python
## Merge both train and test data using pd.concat or append method or you can use merge also 
df = pd.concat([df_train, df_test], axis=0) # axis = 0 means Rows , axis = 1 means columns
df.reset_index(drop=True, inplace=True) # inplace will automatically update into df 
df.head()
```

```python
## Basic Code to understand diffrent types of features 
df.info()
```

```python
## Lets Find Out what is the percentile of this data using describe()
df.describe()
```

```python
## We can Delete that columns which are just waisted not used 
df.drop(['User_ID'], axis = 1, inplace = True) 
# axis = 1 means column wise delete tha data
# inplace = True means after remove USer_ID column it will be auto update into df 
```

```python
df.head()
```

## Check and perform categorical Features

```python
## How many Categorical Features in there 
## Let's Convert this Categorical values in Numerical 
## We can Apply both ways 
# df['Gender'] = pd.get_dummies(df['Gender'], drop_first = 1)
#                 # Or 
df['Gender'] = df['Gender'].map({'F':0, 'M':1})
df.head()
```

```python
## Handle Categorical Feature Age
## Age is also Categorical Feature 
## First Check how many Unique Values in Age  
df['Age'].unique()
```

## In this way is not very Good Practice 
    # pd.get_dummies(df['Age'], drop_first =True)
### Important Note: 
1. Here not to try to do dummies Encoding 
2. Here we will give some ranks and ordinal Encoding let's give some Rank 012345 like
3. bcz If I am training the model my model maths will defiantly will able to understand with respect to the values we are given 
4. 0,1,2,3,4,5 whatever values are we given with respect to the other features my model will definatly to be understandt 
5. this is called as target guiding.
#### we can use also Label Encoding 
#### But here we are using Target Irdinal Encoding using map 

```python
# using Map Function 
# this is called target Ordinal Encoding 
df['Age'] = df['Age'].map({'0-17': 1 , '18-25': 2, '26-35': 3, '36-45': 4, '46-50': 5 , '51-55': 6, '55+': 7})
```

```python
## Fixing Categorical City_Category
# Why i Drop_First A because Two categories is sufficient to represent the Categories 
df_city = pd.get_dummies(df['City_Category'],drop_first = True)
```

```python
df_city.head()
```

```python
## Now Combine the df_city to df dataset 
df = pd.concat([df,df_city], axis = 1)
df.head()
```

```python
## Drop City Category Fature we no need this Feature 
df.drop('City_Category', axis = 1, inplace = True)
```
